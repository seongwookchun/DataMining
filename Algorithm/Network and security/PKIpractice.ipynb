### 1.2 기약잉여계 旣約剩餘界 Reduced Residue System

반대되는 개념 : 완전 잉여계 完全剩餘界 Complete Residue System

법 m에 대한 대표적인 완전잉여계는 {0, 1, 2, ..., m-1}이 있다.  

하지만 완전잉여계가 반드시 위와 같은 꼴인 것만은 아니다. 

예를 들어 법 5에 대한 대표적인 완전잉여계는

A={0,1,2,3,4}인데, 

B={0,1,3,4,7}도 각 원소를 5로 나눈 나머지의 집합을 C라 하면 

C={0,1,3,4,2}로 원소가 중복되지 않으므로 역시 완전잉여계이다.


---

기약잉여계  

완전잉여계에서 서로소인 원소만 모은 집합이다.

````python
# Generate reduced residue system
def gen_RRS(m):
    l_RRS = [1]
    for i in range(1, m):
        if m % i != 0:
            l_RRS.append(i)
    print(l_RRS)

    return l_RRS
````

````python
# m = 15
# a = [1, 2, 4, 7, 8,
#      11, 13, 14]

# m = 16
# a = [1, 3, 5, 7, 9,
#      11, 13, 15]

#a = 5
# for i in range(iteration):
#     print(a**i% m)

m = 20
a = gen_RRS(m)

it_j = 10
#######################################################################
# https://en.wikipedia.org/wiki/Euler%27s_theorem 이해하기 위한 block
for i in range(len(a)):
    print('a[i]', a[i])
    for j in range(it_j):
        
        print(a[i], '^', j, 'mod', m, ':', a[i] ** j % m)
#######################################################################

# phi = 1
# for i in range(len(a)):
#     phi *= a[i]
# print(phi % m)
# print('***')
# for i in range(len(a)):
#     print(phi * (a[i]**len(a)) % m)
````
